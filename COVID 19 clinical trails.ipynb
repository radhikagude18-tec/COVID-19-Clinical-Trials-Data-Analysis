```python
df['Other IDs'] = df['Other IDs'].fillna("None")
```

```python
df['Primary Completion Date'] = df['Primary Completion Date'].fillna("Not Available")
```

```python
df['Start Date'] = df['Start Date'].fillna("Unknown")
```

```python
df['Completion Date'] = df['Completion Date'].fillna("Not Available")
```

```python
df = df.drop(columns=['Results First Posted'])
```

```python
df = df.drop(columns=['Study Documents'])
```
